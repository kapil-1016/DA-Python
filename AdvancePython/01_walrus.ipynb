:= Allows you to assign a value to a variable as part of an expression- so we can both use and save it at a same time.

```python
#normal method
line = input("enter your text: ")
while(line!="quit"):
    print(f"you have entered: {line}")
    line = input("Enter text: ")
```

```python
#using walrus := operator
while(li := input("enter your text: ")) != "quit":  
    print(f"you have entered: {li}")
```
