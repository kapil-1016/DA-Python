Since python is dynamically typed language- doesn't force us to declare types.


but since Python V3.5+, we can optionally define types TYPE HINTS(type annotations)

```python
num : int = 5
```

```python
#as we declare type int, we can access function for int with dot'.' operator.

def sum(n:int, b:int) -> int: # "->" means, function has int return type.
    return n+b
print(sum(1,3))
```
