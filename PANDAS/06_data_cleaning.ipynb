## Data Cleaning


Process of fixing/removing:--->
                            -incomplet, inrelevant, improper data.
                            -around 75% of work with Pandas is data cleaning.

```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
print(df)
```

#### 1. Drop an irrelevant column

```python
df = df.drop(columns=["Legendary", "Type2"])
print(df)
```

#### 2. Handle Missing Data

```python
df = df.dropna(subset=["Type1"]) #dropna() -> function that removes rows or columns that contain null values NaN.
# here subset is the parameter of dropna function which tells pandas to check for particular column that having NaN.
```

```python
print(df.to_string())
```

#### 3. Fix Inconsistent values

```python
df["Type1"] = df["Type1"].replace({"Grass": "GRAss"})
print(df)
```

#### 4. Standardize text

```python
df["Name"] = df["Name"].str.lower()
print(df)
```

```python
# 5 to remove duplicate 
df = df.drop_duplicates()
print(df)
```
