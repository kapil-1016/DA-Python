Series = 1d labelled array that can hold any data type.

Like a single column in a spreadsheet.

```python
import numpy as np 
import pandas as pd
```

```python
data = [100.1, 110.1, 120.1]
series = pd.Series(data) # .Series() is constructor, makes 1d data-collection and also gives meta data.
print(series)
```

```python
# we are also able give our own labels.
house_num =[316, 317, 319]
srs = pd.Series(house_num, index=["krishna's house", "nakul's house", "dev's house"])
print(srs)
```

```python
print(srs.loc["nakul's house"])
#.loc -> label wise location
```

```python
# iloc -> integer based location
print(srs.iloc[0])
```

```python
data = [100, 200, 155, 150, 170]
series = pd.Series(data, index=["a","b","c","d","e"])
print(series[series>=150]) #filter by values
```

```python
calories = {"day1": 2000, "day2": 1500, "day3":1800}
s = pd.Series(calories)
print(s)
```

```python
print(s.loc["day1"])
s.loc["day3"] += 400
print(s)
```

```python
print(s[s>1500])
```
