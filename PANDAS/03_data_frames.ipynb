## DataFrame

A tabular data structure with rows and columns like a excel spreadsheet. 2d

```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Name": ["Krishna", "Nakul", "Manish"],
    "City": ["YNR", "Jind", "Jind"]
}

df = pd.DataFrame(data, index=["C1","C2","C3"])
# .DataFrame is a constructor as well, it identiesf the input data type and convert into 2d table,ie dataframe.
print(df)
```

```python
print(df.loc["C1"])
print(df.loc["C3"])
```

```python
#adding a new column
df["Job"] = ["DA","AIE","FSD"]
print(df)
```

```python
#adding new rows
new_row = pd.DataFrame([{"Name":"Archit", "City":"Hisar", "Job":"NA"},
                        {"Name":"Hitesh", "City":"Panipat", "Job":"FED"}], index=["C4","C5"])
df = pd.concat([df, new_row])
print(df)
```
