# BarChart

Compare categories of data by representing each category with a bar

```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd
```

```python
categories = ["Grains", "Fruit", "Vegetables","Dairy", "Beverages"]
#list of string don't benefited from numpy arrays.
values = np.array([4,3,2,3,1])

plt.figure(figsize=(4.5,4.5)) # adusting figure size accordingly.
plt.bar(categories, values, color="pink")
plt.tick_params(color="yellow")
plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")
plt.show()

#for horizontal bar char => plt.barh()
```
