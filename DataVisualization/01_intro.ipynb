## Data Visualization

Giving visual context to help identify and infer trends, patterns and outliers in datasets.

Outlier is a data point which is significantly different from most of the other values

Its value stand outs because either it is too high or too low compared to other data.

### Importance of Data Visualization

Data is only useful, only if we learn from it.

Delivers data with clarity, efficiency and effectiveness.

Can identify trends over time, corelations, frequency
