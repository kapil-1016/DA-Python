# ScatterGraph

Shows the relationship between two variables

Helps to find a correlation (-ve , +ve or none)

ex-> study hours vs results

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x1 = np.array([0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8])
y1 = np.array([55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87])

x2 = np.array([0, 2, 2, 2, 3, 4, 5, 5 , 7, 7, 8])
y2 = np.array([35, 40, 55, 58, 69, 73, 74, 100, 82, 95, 97])
plt.scatter(x1,y1, color="brown",
            alpha=0.5, s=100, label="Class A")

plt.scatter(x2,y2, color="skyblue",
            alpha=0.5, s=100, label="Class B")
plt.title("Test Scores")
plt.xlabel("Hours")
plt.ylabel("Score")
plt.legend() 
''' A legend in a plot is the little box (usually on the side or top corner) that explains what each color,
    line, or marker in your plot represents.  '''
plt.show()
```
