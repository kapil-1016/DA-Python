# SubPloting

Means multiple plots(axes) arranged in a singule figure window.

Figure - Entire canvas

Ax - A single plot (subplot)

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
figure, axes = plt.subplots(2,2) #.subplots(row, column)
x = np.array([1,2,3,4,5])
axes[0,0].plot(x,x*2, color="red")
axes[0,0].set_title("x*2")

axes[0,1].plot(x,x**2, color="blue")
axes[0,1].set_title("x**2")

axes[1,0].plot(x, x*3, color="pink")
axes[1,0].set_title("x*3")

axes[1,1].plot(x, x**3, color="orange")
axes[1,1].set_title("x**3")
plt.tight_layout()
plt.show()
```
