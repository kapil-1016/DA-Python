```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([2023, 2024, 2025, 2026])
y = np.array([15, 25, 30, 20])
plt.plot(x,y, marker=".", markersize=23, 
         markerfacecolor="#131a26", markeredgecolor="#131a26",
         linestyle="solid", linewidth = 2.5,
         color="red") #none for no line(linestyle)
#accronyms: -> ms-markersize, mfc-markerfacecolor, mec-markeredgecolor
plt.show()
# some markers => o-circle, *-star, D-diamond, .-tiny dot,  ,->pixel(very thin dot)
```

```python
A = np.array([2000, 2010, 2020, 2030])
B1 = np.array([15, 23, 44, 11])
B2 = np.array([50, 14, 28, 42])
B3 = np.array([33, 22, 11, 44])

line_style = dict(marker=".", markersize=10, 
         markerfacecolor="#131a26", markeredgecolor="#131a26",
         linestyle="solid", linewidth = 2.5
         )
# here we have initialized dictionary as a constructor dict() NO-CONFUSSIONS!!!!!!!!!!!
plt.plot(A,B1,color="yellow", **line_style) # ** -> unpacking the dictionary(line_style)
plt.plot(A,B2, color="blue",**line_style)
plt.plot(A,B3,**line_style)
plt.show()
```

# LABELS

```python
#Labels are the text descriptions that appear:
#on the X-axis and Y-axis
#as the title of the graph

#in the legend (to describe multiple lines)
#They help anyone reading your graph understand whats being plotted.
```

```python
X = np.array([2023, 2024, 2025, 2026])
Y1 = np.array([15, 25, 30, 20])
Y2 = np.array([17, 23, 38, 5])
Y3 = np.array([13, 15, 20, 30])

plt.tick_params(axis="both", colors="#4f0dde")
plt.xticks(X) # .xticks() is used to get or set the tick locations and labels of the x-axis in a plot.

plt.title("Calss size", fontsize=20,
          fontweight="bold", color="#384d04")

plt.xlabel("Year", fontsize=20,
           fontweight="bold",color="#d2f57d")
plt.ylabel("No. of Students", fontsize=20,
           fontweight="bold",color="#d2f57d")

plt.plot(X,Y1)
plt.plot(X,Y2)
plt.plot(X,Y3)
plt.show()
```

## Grid

Helps make plots easier to read by adding reference lines, especially when comparing data points.

```python
n=[1,2,3,4,5]
m=[5,10,15,20,25]
plt.grid(axis="both",color="pink")
plt.plot(n,m)
plt.show()
```
