# Histogram

```python
# Visual representation of the distribuition of quantitative data.
#  They group values into bins (intervals) and counts how many fall in each range.
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
scores = np.random.normal(loc=70, scale=10, size=100) #.normal -> normal distribution, loc->mean, scale->std,size->no. of data pts.
# std -> shows how far is our data from center.
scores = np.clip(scores, 0, 100) # clip-> changes number which is less min to min and which is greater than max to max.
plt.hist(scores, bins=10, color="lightgreen", edgecolor="black")
plt.title("Exam scores")
plt.xlabel("Score")
plt.ylabel("No of students")
plt.show()
```
