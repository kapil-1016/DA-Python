```python
# Write a class “Calculator” capable of finding square, cube and square root of a number.
class Calculator:
    def __init__(self, n):
        self.n = n
    
    def square(self):
        print(f"The square is: {self.n*self.n}")
    
    def cube(self):
        print(f"Cube is: {self.n*self.n*self.n}")
    
    def sqroot(self):
        print(f"Square root is:{self.n**1/2}")


sq = Calculator(4)
sq.square()

cb = Calculator(3)
cb.cube()

a = Calculator(16)
a.sqroot()
```
