```python
# Create a class “Programmer” for storing information of few programmers working at Microsoft.
```

```python
class Programmer:
    company = "Microsoft"
    def __init__(self, name, location, shift):
        self.name = name
        self.location = location
        self.shift = shift

krishna = Programmer("Krishna Kapil", "Gurugram", "Day")
nakul = Programmer("Nakul Jain", "Tokyo", "Night")
manish = Programmer("Manish Malik", "Thailand", "Flexible")
sushant = Programmer("Sushant Tiwari", "Gurugram", "Day")

print(sushant.shift, sushant.name)
print(krishna.name, krishna.company, krishna.location)
print(nakul.name, nakul.company, nakul.location, nakul.shift)
```
