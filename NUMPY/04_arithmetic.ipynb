```python
import numpy as np
```

```python
# scaler arithmetic, normal python method.
arr = np.array([10,20,30,40])
print(arr + 10)
print(arr * 10)
print(arr ** 2)
```

### vectorized math functions: 

numpy allows vectorized math functions, means-> operations are done on entire array at a same time.

```python
radii = np.array([1,2,3])
print(np.pi * radii**2)
```

```python
# Element wise arithmetic!
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

print(arr1 + arr2) #[5,7,9]
```

```python
#comparison operators!

marks = np.array([55, 100, 69, 69, 10])
print(marks>=30)
```
