### Filtering

Refers to the process of selecting elements from a given array according a specific condition.

```python
ages = np.array([[21, 17, 19, 20, 16, 30, 18, 65],
                 [39, 22, 15, 99, 18, 19, 20, 69]])
teenageres = ages[ages<=18]
adults = ages[(ages>=18) & (ages<=65)]
print(teenageres)
print(adults)
```

```python
ages = np.array([[21, 17, 19, 20, 16, 30, 18, 65],
                 [39, 22, 15, 99, 18, 19, 20, 69]])
# where function syntax  -> where(condition, value_if_true, value_if_false)

adults = np.where(ages >= 18 , ages, 0)
print(adults)
```
