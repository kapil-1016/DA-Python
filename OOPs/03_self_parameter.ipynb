'self' refers to the CURRENT object of the class

```python
class Student:
    name = "krisna"
    roll_no = 1812

    @staticmethod #no need of self, bcz of now greet is a static metho
    def greet():
        print("Hello ")

kk = Student()
# Student.greet(kk) #this is done bts by python automatically, when we use self parameter.
kk.greet()
```

@staticmethod: inside a python class that:
                -> does not use self(object reference)
                -> does not use cls(class reference)
                -> works independently-it belongs to class basically, but doesn't need object data.
we declare it like - " @staticmethod "
