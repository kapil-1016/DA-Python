### SINGLE INHERITANCE

When a derived class only inherits one base class

```python
class Employee: #base class
    company = "ITC"
    def show(self):
        print(f"name of employee is : {self.name} and salary is:{self.salary}")

class Programmer(Employee): # programmer is derived class or child class and employee is the parent or base class. 
    #company = "ITC Infotech"
    def ShowLanguage(self):
        print(f"name is :{self.name} and he is good with {self.language} language.")

a = Employee()
b = Programmer()
print(a.company, b.company)
    
```

### Multiple Inheritance

When a child class inherits from more than one parent class.

```python
class Mother:
    def hobby(self):
        print("i love cooking")

class Father:
    def skill(self):
        print("i can drive a car")

class Son(Mother, Father):
    def liking(self):
        print("I have keen interest in powerlifting.")

child = Son()
child.hobby()
child.skill()
child.liking()
```

### Multilevel Inheritance

When a child class becomes a parent class for other child class.

parent -> child1 -> child2

```python
class Grandfather:
    def house(self):
        print("Grandfather: I built a house.")

class Father(Grandfather):
    def car(self):
        print("Father: I bought a car.")

class Son(Father):
    def bike(self):
        print("Son: I have a bike.")

# create object of Son
s = Son()
s.house()   # from Grandfather
s.car()     # from Father
s.bike()    # from Son
```
