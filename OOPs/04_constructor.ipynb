CONSTRUCTOR: A special method that automatically runs when an object of class is create.
-> In python, constructor method => __init__()

__ init__(): is a special method(constructor) in a class that runs automatically when object is created.
                It is used to setup object's data.
                It take 'self' argument and can take more.

```python
class Employee:
    #language = "python"
    #salary = 100000
    def __init__(self, name, salary, language): #dunder method starts with __
        self.name = name
        self.salary = salary
        self.language = language
        print("object is created.")

krishna = Employee("krishna", "200000","Python")
print(krishna.language)
kumar = Employee("kumar", "100000", "Java")
print(kumar.language)
```
