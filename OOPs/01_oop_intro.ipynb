OOPs: A way of writting code where we organize our codes into objects.
OBJECTS occupies: Data(attributes) and Behaviour(functions)

Why OOPs: 
            -> Organized(we divide larger problems in form of classes.)
            -> Reusable(once class is created, we can use it anytime).
            -> Scalable(good for bigger projects).
            -> Close to real world thinking.

CLASS: Kinda blueprint for creating objects.(class should be written in PASCAL case)
OBJECT: One of the instances of class.

CLASS ATTRIBUTE: An attribute which belongs to a class not to a particular object.  
INSTANCE ATTRIBUITE: An attribute which belongs to an object.

```python
class Employee:
    language = "python"
    salary = 700000

krishna = Employee()
krishna.name = "Krishna Kapil" #object "krishna's" attribute
print(krishna.name, krishna.language, krishna.salary)
# here language and salary both are class "empoyee's" attributes not krishna's.

mirza = Employee()
mirza.name = "Md Mirza Ahmed" #object "mirza's" attribute
print(mirza.name, mirza.salary, mirza.language)
```
